A의 이동 가능성은 L(왼쪽으로 1칸 이동),  R(오른쪽으로 1칸 이동),  U(위쪽으로 1칸 이동), D(아래쪽으로 1칸 이동) 이렇게 4가지 방향만 이동이 가능함. 시작 좌표는 항상 (1,1)이며, N by N의 공간을 벗어나게 된다면, 그 움직임은 무시가 된다. 예: (1,1)에 있는데, U이 나오면 움직임을 하지 않고 무시한다.

### 생각의 출발점

1. 입력 : 우선 입력 예시를 기준으로 생각을 하자. 일단 5라고 하면, 5 by 5의 정사각형 공간에서 이루어지는 것이고, 내가 10을 입력을 했다고 생각을 하면 10 by 10의 정사각형의 공간에서 이루어진다는 것을 생각을 할 수 있다. → 입력의 크기를 받으면 되니, input을 사용을 하고 & input은 숫자를 입력을 해도 문자열이니 숫자로 강제형변환 정도 생각이 나면 됨.
    
    
2. 이동에 대한  R, L ,U ,D 를 어떻게 표현을 할까? → 이 부분이 이 문제를 푸는 가장 기본적인 출발점이 된다!!!
    
    주어진 상황을 잘 보면 우리가 익숙하게 아는 좌표 평면으로 되어 있으며, 이에 대한 것으로 표현을 하는것이 쉽다. 다만 이것을 처리하기 위해서는 X, Y라고 각기 독립적으로 처리를 해야하는 점이 있다. → 그러나 아래에서 출력에서 X,Y라고 각기 힌트를 주었다. 
    
    기존 좌표가 (a,b) 라고 하면, R 이 나오면 (a+1, b)이런 식이 되고, L 이 나오면 (a-1, b)가 되고, U가 나오면 (a, b+1), D가 나오면 (a, b-1)이 된다. 이것은 일반적인 좌표평면이고, 주어진 문제의 상황(맨 왼쪽 상단이 1,1 → 맨 오른쪽 하단이 N,N) 에 맞게 다시 수정을 하면, R이 나오면 (a, b+1) ,  L이 나오면(a, b-1), U이 나오면 (a-1, b), D가 나오면(a+1, b)로 수정하면 된다. 
    
    이러한 부분에 대해서 정리를 하면 된다.
    
    방법1) 교재의 방식 :  이동하는 타입 지정 , 이동하는 타입 별로 x, y에 대한 변화
    
    방법2) 주어진 이동에 대한 파이썬의  dict로 구현 →  키 : 이동하는 타입, 값 :[ 이동 타입별 x 이동, 이동 타입별 y이동] → 하나의 dict에 다 구현을 하는 방식을 사용하는 방법
    
3.  주어진 값들에 대해서 하나씩 처리하기 → for 문으로 주어진 이동 규칙을 하나씩 하자!
4. 놓칠 수 있는 중요한 부분!!! 범위를 벗어나는 경우에 대한 이동을 하지 않는 다는 점!!!
    
    예로 나타나 있던 상황으로 문제에서 충분히 이러한 예외 상황에 대한 사례와 이동 시 예를 주었음. 따라서 이러한 부분에 대해서 미리 생각할 수 있는 기회를 주고 있음!
    
    → 아이디어 : 범위를 벗어난다는 것은 계산한 결과과 주어진 판의 밖으로 나가면 이동하지 않는다!!
    
    **→ 구현 : x, y를 기준으로 바라보고 있으니, x,y가 주어진 범위 밖으로 된다면 → x가 1보다 작거나, n보다 큰 경우에는 위/아래로 주어진 범위를 벗어나게 되고, y가 1보다 작거나 n보다 큰 경우에는 좌/우로 주어진 범위를 멋어나게 된다.!!!!!!! → if로 각기 경우에 대한 예외처리를 하면 된다!! + 이 경우에 아무것도 안 해야하니!!!!**
    
5. 최종 출력 : 최종적으로 남은  x , y 의 값을 출력!

```python
# 내 풀이
N = int(input())
rlud = input().split()
start = [1, 1]
for i in rlud :
    if i == 'R' and start[1] < N :
        start[1] += 1
    elif i == 'L' and start[1] > 1 :
        start[1] -= 1
    elif i == 'U' and start[0] > 1 :
        start[0] -= 1
    elif i == 'D' and start[0] < N :
        start[0] += 1

print(start[0], start[1])
```

```python
# 교재 방식

# 입력 관련 부분
n = 5
x, y = 1, 1
plans = ["R","R","R","U","D","D"]


# L/R/U/D의 이동관련 부분 설정
dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]
move_types = ["L","R","U","D"]

# 이동 계획을 하나씩 돌면서 처리!!-> for
for plan in plans :
	# 이동 후 좌표 구하기
	for i in range(len(move_types)):
		if plan == move_types[i]:
			nx = x + dx[i]
			ny = y + dy[i]
		
	# 공간을 벗어나는 경우
	if nx <1 or ny <1 or nx >n or ny > n:
		continue
	
	# 이동 수행
	x, y = nx, ny

print(x,y)
```

```python
# 고정된 입력으로 (Dictionary)
n = 5
x = 1
y = 1
plans = ["R","R","R","U","D", "D"]
move = {"R":[0,1], "L" :[0,-1], "U":[-1, 0], "D":[1,0]}

for m in plans:
    dx, dy = move[m]
    if x +dx <1 or x + dx > n or y + dy <1 or y + dy > n:
        continue
    else:
        x = x+dx
        y = y+dy
print(x,y)
```

### 구현

- 사전 필요 사항 : 가로에 대해서는 알파벳으로 구현이 되어 있음!!!

<aside>
💡 ord ( ) 함수 : 특정한 문자를 아스키 코드로 변환을 해주는 함수
chr ( ) 함수 : 아스키 코드값을 문자로 변환해주는 함수

<aside>
💡 이 문제의 핵심은 모든 정보 및 대상들을 대상으로 수치화 시켜서, 이것들에 대한 수치적인 규칙으로 환산을 하여 접근을 한다는 점!!!

</aside>

```python
# 내 풀이
row = list('12345678')
column = list('abcdefgh')

# 주어진 위치에서 이동이 가능한 8가지 이동에 대한 정의!!!!
steps = [(-2, -1), (-2, 1), (2, 1), (2, -1),
        (1, 2), (1, -2), (-1, 2), (-1, -2)]

v, h = list(input())
n = 8
for i in steps :
    if chr(ord(v) + i[0]) not in column :
        n -=1
    elif str(int(h) + i[1]) not in row :
        n -=1

print(n)
```

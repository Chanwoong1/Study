## 기본사항

- 재귀함수 Recursive Function : 자기 자신을 다시 호출하는 함수를 의미함.
- 아래의 코드를 해보자... 그러면 아래와 같이 쭉 실행을 하다가 마지막에서 시행횟수가 최대를 초과했다가 나오면서 에러가 발생을 한다. 이유는 파이썬의 경우에서는 인터프리터가 호출하는데 제한 횟수가 있어서 아래와 같은 현상이 나타나고, 무한정 부를 수 없다. → 재귀함수를 할 때는 주의가 필요함!!!!

```python
def recursive_func():
    print("나는 재귀함수다!!")
    recursive_func()

# 재귀 함수 실행
recursive_func()
```

## Factorial 함수에 재귀함수 사용

- 아래 그림과 같이 factorial 함수에 대한 내용이며, 그 아래 그림은 factorial  함수에 대해서 재귀함수적인 관점으로 본 그림임.

```python
def my_factorial(n):
		# 진짜 불러지는지 확인을 위해서 print 함수를 사용을 함..
    print("Factorial n값은?" + str(n))
    if n <=1 :
        return 1
    else:
        return  n * my_factorial(n - 1)

### 실제 4!에 대한 계산 시행
my_factorial(4)
```

### 참고) factorial  함수의 다른 구현 방법들

- math  패키지 사용


```python
import math

n =4
print(math.factorial(int(n)))
```

- 반복문 사용

```python
# While 문 사용
n = 4
fact = 1 # 계산에 대한 최종 결과 처리용
a=1 # 증감용
while a <= n :
    fact*=a
    a+=1
print( str(n)+"! =" + str(fact))



##############################
# for 문 사용
n = 4
fact = 1
for i in range(1, n+1):
    fact  *= i
print( str(n)+"! =" + str(fact))
```

- 아래의 Factorial의 정의부터 다시 알고리즘의 관점으로 접근해보자.

$$\text{Factorial of N} = factorial(N) = N! = N * (N-1) * \cdots * 2 * 1$$

위의 경우에 있어서는 입력의 크기가 상당히 커지게 되면, 그 값이 커지는 속도가 상당하다.
하지만 알고리즘 및 시간복잡도의 관점에서는 T(N)의 관점이 된다.
입력 크기 n에 대해서 함수의 수행시간이 T(n)이라고 하고, 함수에서  n * f(n-1)의 연산에 소비되는 시간을 c라고 한다면, T(n) = T(n-1) + c 가 될 것이다.
이 식을 전개를 해본다면
T(n) = T(n-1) + c
        = T(n-2) + c + c = T(n-2) + 2c
        = T(n-3) + c + c + c = T(n-3) + 3c
        = ...
        = T(1) + c + c+ ... + c = T(1) + (n-1)c
        ≤ c + (n-1) c = c + nc - c = nc 
따라서 시간 복잡도는 O(n)이 된다!!!!!

참고로  for문의 경우에는 그냥 갯수만큼 증가이니 앞에서 본 것과 같이 시간복잡도는 O(n)이 된다!!!

- 다시 앞에서 한 Factorial 문제에 대해서 재귀함수 version vs 반복문 version에 대해서 생각해보자. 아래 코드는 위에서 한 것들을 다시 가져온 것임.

```python
n = 4
fact = 1
for i in range(1, n+1):
    fact  *= i
print( str(n)+"! =" + str(fact))



##########################################################
def my_factorial(n):
		# 진짜 불러지는지 확인을 위해서 print 함수를 사용을 함..
    print("Factorial n값은?" + str(n))
    if n <=1 :
        return 1
    else:
        return  n * my_factorial(n - 1)

### 실제 4!에 대한 계산 시행
my_factorial(4)
```

- 위의 코드를 보고 재귀함수를 이용한 것들의 특징은
- 구현이 간단하다(factorial 의 점화식을 그대로 구현을 하면 됨) → 생각이 필요한 것이 아니라 그냥 단순히 주어진 수식을 그대로 코딩하면 되는 특징 → 뒤에서 Dynamic programming  에서 다시 나타날 듯..

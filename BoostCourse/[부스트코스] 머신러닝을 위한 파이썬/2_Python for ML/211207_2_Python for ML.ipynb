# Housing data 불러오기

```python
import pandas as pd
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'
df_data = pd.read_csv(data_url, sep = '\s+', header = None) # csv 타임 데이터 로드, sep은 빈공간으로 지정
```

```python
df_data.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df_data.head()
```

```python
# X0 추가, Y값 없애기
df_data['weight_0'] = 1
df_data = df_data.drop('MEDV', axis = 1)
df_data.head()
```

```python
# Dataframe -> numpy, 곱셈연산
import numpy as np
df_matrix = df_data.values
weight_vector = np.random.random_sample((14, 1))
```

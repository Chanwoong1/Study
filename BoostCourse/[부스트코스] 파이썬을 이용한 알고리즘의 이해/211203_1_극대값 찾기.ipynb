# 2차원의 경우
n개의 행과 m개의 열이 있다고 하자.

먼저 극대값이 무엇인지 정의해야한다.

2차원에서의 극대값은 언덕이다.

사방의 수보다 같거나 커야하기 때문이다.

## 탐욕 상승 알고리즘
방향을 고르고, 그 방향을 따라 극대값을 찾는다.

2차원에서의 최악의 경우 시간복잡도는 O(nm) 을 갖는다. 만약 n = m 인 경우 O(n^2)

## 2차원에서의 이진 탐색

수 많은 열 중 하나인 j열을 골라서 1차원의 극대값을 구한다(j = m/2)인 경우가 가장 효율적. 그 점이 (i, j) 라고 한다면, 그 다음에는 i행에서 극대값을 찾는다.

-> 이 방법은 정확하지 않다. i행에서 j열에 맞는 극대값이 존재하지 않을수도 있기 때문이다.

## 그러면 실제로 작동하는 알고리즘은 무엇일까 ?

복잡도의 관점에서 탐욕 상승 알고리즘보다 재귀를 이용하는 알고리즘이 효율적이다.

j열을 고르고, 그 열에서 최대값을 구한다. 그 값이 (i, j)라면 (i, j-1), (i, j), (i, j+1)을 비교한다.

만약  (i, j-1) > (i, j) 이면 왼쪽 열을 고른다. 반대의 경우 오른쪽. 이 작업을 반복한다. 이 두 경우가 만족하지않으면, 그 점이 극대값이다.

이 경우 극대값을 찾는데, 열 개수가 절반으로 줄어들게 된다. 따라서 시간 복잡도도 줄어들게 된다.

이 작업을 반복하면 최악의 경우 하나의 열이 남게 되는데, 그 때, 1차원의 최대값을 찾으면 된다.

시간 복잡도는

T(n, m) = T(n, m/2) + O(n) = O(n) + O(n) + ... + O(n) (log2m번)  = O(nlog2m)

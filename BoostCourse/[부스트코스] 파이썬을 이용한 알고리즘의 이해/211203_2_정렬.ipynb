# 정렬
왜 정렬을 해야 할까?
- 확실한 응용 사례
    - MP3 보관함 정렬
    - 전화번호부 정리
- 일단 정렬되면 쉬워지는 문제들
    -  중간값 또는 가장 가까운 쌍 찾기
    -  이진 탐색, 통계적 이상치 확인
-  생소한 응용 사례
    -  데이터 압축: 정렬로 중복된 부분 검출
    -  컴퓨터 그래픽: 앞에서 뒤로 화면 렌더링
    
## 삽입 정렬
삽입 정렬 (A, n) A[1 . . n]

j ← 2 부터 n 까지 반복

알맞은 자리로 쌍별 스왑을 통해서

키 A[j] 를 (이미 정렬된) 보조 배열인 A[1 .. j-1] 에 삽입

실행 시간은? Θ(n^2)임. 왜냐하면 Θ(n^2) 비교와 Θ(n^2) 스왑이므로.
예시: 입력값이 A = [n, n − 1, n − 2, . . . , 2, 1]인 경우


## 이진 삽입 정렬

이진 삽입 정렬 (A, n)

j ← 2 부터 n 까지 반복

A[1 . . n]

(이미 정렬된) 보조 배열인 A[1 .. j-1]에 키 A[j] 삽입.

알맞은 위치를 찾기 위해 이진 탐색 사용.

이진 탐색은 Θ(log n) 시간이 걸림.
하지만 삽입 후 항목을 옮기는 것도 Θ(n) 시간이 걸림.

복잡도: Θ(n log n) 비교

(n^2) 스왑

## 합병 정렬이란?

합병 정렬 A[1 . . n]

분할 정복
1. n = 1일 때, 바로 끝 (정렬할 거 없음).
 
2. 그 외의 경우, A[ 1 . . n/2 ]과 A[ n/2+1 . . n ]을 재귀적으로 정렬.

3. 두 정렬된 보조 배열 “합병”

총 n 개의 항목을 합병하는 데 걸리는 시간 = Θ(n) (선형적 시간)
